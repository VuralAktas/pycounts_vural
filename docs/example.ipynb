# Example usage

To use `pycounts_vural` in a project:

```python
import pycounts_vural

print(pycounts_vural.__version__)
```

```python
from pycounts_vural.pycounts_vural import count_words
from pycounts_vural.plotting import plot_words
```

```python
quote = "Insanity is doing the same thing over and over and expecting different results."
with open("einstein.txt", "w") as file:
    file.write(quote)
```

```python
counts = count_words("einstein.txt")
print(counts)
```

```python
fig = plot_words(counts, n=5)
```
